```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregue o dataset
df = pd.read_csv('./assets/Unicorn_Companies.csv')

# Visualize as primeiras linhas do dataset
print(df.head())

# Verifique as colunas disponíveis
print(df.columns)

# Configurações gerais de estilo
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))
```

```python
# 1. Empresas unicórnios por país
unicornios_por_pais = df['Country'].value_counts()
plt.figure(figsize=(12, 8))
ax1 = sns.barplot(y=unicornios_por_pais.index, x=unicornios_por_pais.values, palette='viridis')
ax1.set_title('Número de Empresas Unicórnios por País', fontsize=16)
ax1.set_xlabel('Número de Unicórnios', fontsize=14)
ax1.set_ylabel('País', fontsize=14)
for i in ax1.containers:
    ax1.bar_label(i, fmt='%d', fontsize=12)
plt.show()
```

```python
# 2. Setores mais comuns entre unicórnios
setores_mais_comuns = df['Industry'].value_counts().head(10)
plt.figure(figsize=(12, 8))
ax2 = sns.barplot(y=setores_mais_comuns.index, x=setores_mais_comuns.values, palette='coolwarm')
ax2.set_title('Top 10 Setores das Empresas Unicórnios', fontsize=16)
ax2.set_xlabel('Número de Empresas', fontsize=14)
ax2.set_ylabel('Setor', fontsize=14)
for i in ax2.containers:
    ax2.bar_label(i, fmt='%d', fontsize=12)
plt.show()
```

```python
# 3. Valor médio dos unicórnios por setor
df['Valuation'] = pd.to_numeric(df['Valuation'], errors='coerce')
valor_medio_setor = df.groupby('Industry')['Valuation'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 8))
ax3 = sns.barplot(y=valor_medio_setor.index, x=valor_medio_setor.values, palette='magma', hue=valor_medio_setor.index, dodge=False)
ax3.set_title('Top 10 Setores com Maior Valor Médio das Empresas Unicórnios', fontsize=16)
ax3.set_xlabel('Valor Médio (em Bilhões)', fontsize=14)
ax3.set_ylabel('Setor', fontsize=14)
if ax3.legend_ is not None:
    ax3.legend_.remove()
for i in ax3.containers:
    ax3.bar_label(i, fmt='%.2f', fontsize=12)
plt.show()
```
